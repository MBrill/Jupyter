# RISE und Widgets
Wir können RISE auch mit Widgets aus dem Paket *ipywidgets' kombinieren, was interaktive Folien erzeugt.

```python
from ipywidgets import interact

def my_function(x):
    return x

# create a slider
interact(my_function, x=20)
```
