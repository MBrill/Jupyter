# Ebene Parameterkurven - Aufgabe 4
In der Aufgabe 4 des Übungsblatts sbestimmen wir einen Parameterwert einer Kurve, so dass die Tangente an diesem Punkt der Spur parallel zu einer 
gegebenen Linie ist. Die Berechnungen dazu finden wir in der Musterlösung. Hier geben wir die Parameterkurve und die Linie grafisch aus.

Die Linie können wir eine Parameterkurve behandeln. Wir benötigen jedoch nur zwei Punkte für die Ausgabe.

```python
import numpy as np

def curvePoint(t):
    return t*(t*t + 4.0), 6.0*t*t


def curve(a=-1.75, b=1.75, n=100):
    t = np.linspace(a, b, n)
    return curvePoint(t)


def linePoint(t):
    return 7.0*t, 10.0*t - 5.0


def line(a=0.0, b=2.0):
    t = np.array([a, b])
    return linePoint(t)

```

## Ausgabe der Kurve und der Linie
Wir berechnen die Koordinaten und geben die beiden Spuren aus.

```python
import matplotlib.pyplot as plt

x, y = curve()
lx, ly = line()

fig = plt.figure()
plt.plot(x, y, 'g-')
plt.plot(lx, ly, 'b-')
plt.title('Spur der Parameterkurve $(t^3+4t, 6t^2)$ mit Parameterintervall $[-1.75, 1.75]$ und der Linie $(7t, 12t-5)$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()

plt.show() 
```
