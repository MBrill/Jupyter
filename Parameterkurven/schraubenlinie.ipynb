# Ausgabe der Schraubenlinie in Plotly

## Schraubenlinie über der logarithmischen Spirale
Wir können statt dem Kreis wie bei der Helix auch eine logarithmische Spirale verwenden und eine Schraubenlinie erzeugen.
Dazu verwenden wir die Funktion *logSpiralHelix* im Modul *space_curves*. 
Dabei verwenden wir die Defaultwerte a=1 und b=0.08 für die Parameter der Spirale, also Ganghöhe verwenden wir 1.
Das Parameterintervall ist [0, 6pi] und wir verwenden 200 Punkte für den Polygonzug.

Der restliche Ablauf verläuft analog
zur Ausgabe der Helix. 

```python
import plotly.graph_objects as go
import space_curves as sp


x, y, z = sp.logSpiralHelix()

logspiral = go.Scatter3d(x=x, y=y, z=z,
                     name='Logarithmische Spirale',
                     mode='lines',
                     line = dict(color='green', width=8),
                     showlegend=False)

layout = go.Layout(scene_xaxis_visible=True, 
                   scene_yaxis_visible=True, 
                   scene_zaxis_visible=True,
                   showlegend=False)

fig =  go.Figure(data=[logspiral], layout=layout)

fig.update_layout(title='Logarithmische Spirale',
                  title_x=0.5,
                  scene_camera_eye_z=1.55,
                  autosize=False,
                  width=800,
                  height=800,
                  showlegend=False)
                  
fig.show()
```
